```python
spambase_df = pd.read_csv('spambase.csv')
spambase_df
```

```python
testData=spambase_df.get(['Spam'])
```

```python
X = spambase_df.drop(columns=['Spam'])
y = spambase_df['Spam']
train_X, valid_X, train_y, valid_y = train_test_split(X, y, test_size=0.33, random_state=1)
```
